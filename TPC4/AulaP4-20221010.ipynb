### Exercício 1 (repescado): estaOrdenada
- Defina uma função que recebe uma lista de números e verifica se a mesma está ordenada por ordem crescente dando como resultado um valor **True** ou **False**.

````python
# Solução do exercício 1
def estaOrdenada(lista):
    ordenada == True
    i = 1
    while ordenada and i < len(lista):
        if lista[i] < lista[i-1]:
            ordenada = False 
    return ordenada 
````

````python
#exemplo
def ordenaLista2(lista):
    trocas = True
    while trocas:
        i = 1
        trocas = False
        while i < len(lista):
            if lista[i] < lista[i-1]:
                lista[i], lista[i-1] = lista[i-1], lista[i]
                trocas = True
            i = i + 1
    return lista
print(ordenaLista2([2, 7, 1, 6, 8])) 
````

````python
import time
cinema = []

def inserirsala(cinema):
    filme = input("Introduza o nome do filme:")
    for s in cinema:
        nlugares, vendidos, nome = s
        if nome == filme:
            print("o filme já existe")
            menu()
    nlugares = int(input("Introduza o número de lugares da sala:"))
    vendidos = []
    cinema.append((nlugares, vendidos, filme))
    print("A sala foi criada!")
    print("salas:", cinema)
    menu()

def removersala(cinema):
    index = int(input("Introduza o indice da sala:"))
    cinema.pop(index)
    print("Sala removida!")
    print("Salas:", cinema)
    menu()

#listar = listarSalas
def listarSalas(cinema):
    print("sala    filme")
    print("-------------------")
    for p in cinema:
        nlugares, vendidos, filme = p
        print("Sala#{0}".format(cinema.index(p)), " ", filme)
    time.sleep(1)
    menu()

#disponivel = lugarDisponivel
def lugarDisponivel(cinema, filme, lugar):
    for p in cinema:
        nlugares, vendidos, nome = p
        if filme == nome:
            if lugar in vendidos:
                res = False
            else:
                res = True
        else:
            res = False
    return res

def vendebilhete(cinema, filme, lugar):
    if lugarDisponivel(cinema, filme, lugar):
        for p in cinema:
            nlugares, vendidos, nome = p
            if filme == nome:
                vendidos.append(lugar)
        print("Lugar vendido!")
        print("Salas", cinema)
    else:
        print("Lugar indisponivel!")
    menu()

def listardisponibilidadedesala(cinema):
    print("Sala    Filme    Lugares disponiveis")
    print("------------------")
    for p in cinema:
        nlugares, vendidos, filme = p
        print("Sala#{0}".format(cinema.index(p)), " ", filme, " ", nlugares - len(vendidos))
    time.sleep(1)
    menu()

def menu():
    print("""
    (1) Criar sala
    (2) Remover sala
    (3) Listar salas
    (4) Lugar disponivel
    (5) Vender bilhete
    (6) Consultar salas
    (0) Sair/reset
    """)
    menuOption = int(input("Escolha uma funcao:"))
    if menuOption == 1:
        inserirsala(cinema)
    elif menuOption == 2:
        removersala(cinema)
    elif menuOption == 3:
        listarSalas(cinema)
    elif menuOption == 4:
        filme = input("Introduza o nome do filme:")
        lugar = int(input("Introduza o lugar:"))
        print("Lugar disponivel?", lugarDisponivel(cinema, filme, lugar))
    elif menuOption == 5:
        filme = input("Introduza o nome do filme:")
        lugar = int(input("Introduza o lugar:"))
        vendebilhete(cinema, filme, lugar)
    elif menuOption == 6:
        listardisponibilidadedesala(cinema)
    elif menuOption == 0:
        print("Saíada efetuada!")
        print("Salas:", cinema)
        
menu()



````
